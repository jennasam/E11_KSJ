```python
#3.1
Al_1 = pd.read_csv("Shielded_1Al.csv")
x1 = Al_1["Counts per 10 Seconds"]

Al_2 = pd.read_csv("Shielded_2Al.csv")
x2 =  Al_2["Counts per 10 Seconds"]

Al_3 = pd.read_csv("Shielded_3Al.csv")
x3 =  Al_3["Counts per 10 Seconds"]

Tg_1 = pd.read_csv("Shielded_Tungsten.csv")
y1 = Tg_1["Counts per 10 Seconds"]

Tg_2 = pd.read_csv("Shielded_Tungsten2.csv")
y2 = Tg_2["Counts per 10 Seconds"]

Tg_3 = pd.read_csv("Shielded_Tungsten3.csv")
y3 = Tg_3["Counts per 10 Seconds"]
```

```python
#3.2a

th_unc = np.sqrt(len(Al_1))
print("The theoretical uncertainty is", th_unc,)
Al_1m = np.mean(x1)
Al_1std = np.std(x1)

print("The mean of counts per 10 seconds for 1 Aluminum sheet is", Al_1m, "with a std of", Al_1std, ".\nThe difference between the std and theoretical uncertainty is", th_unc-Al_1std)


Al_2m = np.mean(x2)
Al_2std = np.std(x2)
print("the mean of counts per 10 seconds for 2 Aluminum sheets is", Al_2m, "with a std of", Al_2std, ".\nThe difference between the std and theoretical uncertainty is", th_unc-Al_2std)

Al_3m = np.mean(x3)
Al_3std = np.std(x3)
print("the mean of counts per 10 seconds for 3 Aluminum sheets is", Al_3m, "with a std of", Al_3std , ".\nThe difference between the std and""theoretical uncertainty is", th_unc-Al_3std)


Tung_1m = np.mean(y1)
Tung_1std = np.std(y1)
print("the mean of counts per 10 seconds for 1 Tungsten sheet is", Tung_1m, "with a std of", Tung_1std, ".\nThe difference between the std and theoretical uncertainty is", th_unc-Tung_1std)

Tung_2m = np.mean(y2)
Tung_2std = np.std(y2)
print("the mean of counts per 10 seconds for 2 Tungsten sheets is", Tung_2m, "with a std of", Tung_2std, ".\nThe difference between the std and theoretical uncertainty is", th_unc-Tung_2std)


Tung_3m = np.mean(y3)
Tung_3std = np.std(y3)
print("the mean of counts per 10 seconds for 3 Tungsten sheets is", Tung_3m, "with a std of", Tung_3std, ".\nThe difference between the std and theoretical uncertainty is", th_unc-Tung_3std)
```

3.2b. As the thickness of the absorber material increases, the counts decrease. This is a result of there being a higher change of interaction when more material is present since it is more dense. Counts and material thickness are inversely proportional.

```python
#3.3
tc_Al_1 = np.sum(x1) 
tc_Al_2 = np.sum(x2)
tc_Al_3 = np.sum(x3)

plt.errorbar([1, 1.75, 2.5], [tc_Al_1, tc_Al_2,tc_Al_3 ], yerr = th_unc)
plt.yscale("log")
plt.ylabel('total counts')
plt.xlabel("absorber thickness (cm)")
plt.show()
```

```python
tc_Tg_1 = np.sum(y1) 
tc_Tg_2 = np.sum(y2)
tc_Tg_3 = np.sum(y3)

plt.errorbar([1, 2, 3], [tc_Tg_1, tc_Tg_2,tc_Tg_3 ], yerr= th_unc) 
plt.yscale("log")
plt.ylabel('total counts')
plt.xlabel("absorber thickness (cm)")
```

```python
#3.4a
Al_coef = (np.log(tc_Al_2) - np.log(tc_Al_3))/ (tc_Al_2 - tc_Al_3)
Tg_coef = (np.log(tc_Tg_1) - np.log(tc_Tg_2))/ (tc_Tg_1 - tc_Tg_2)

print("The absorption coefficient for Aluminum is", Al_coef, "and the absorption coefficient of Tungsten is", Tg_coef)
```
