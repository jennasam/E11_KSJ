```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np
import math
from scipy.stats import norm
```

```python
data = pd.read_csv("Loc1_session1_group1_2025-04-18_15-39-44.csv")
time_data = data["Time"]
rad_data = data["counts"]
temp_data = data["Temperature"]
PM25_data = data["PM25"]
```

```python
plt.plot(time_data, rad_data)
plt.xlabel("Time (sec)")
plt.ylabel("Radiation Data (CPM)")
plt.title('Radiation Data vs Time')
```

```python
rad_mean = np.mean(np.array(rad_data))
rad_std = np.std(rad_data)
print(f"The mean is {rad_mean} and the std is {rad_std}")
```

```python
rad_sigma = np.sqrt(rad_mean)
print(f"The variance is {rad_sigma}. The measured standard deviation is comparable to the expected variance, so we can apply counting statistics to our data")
```

```python
N = len(rad_data)
N_samples = 5

rad_uncert = (math.sqrt(N)) / (math.sqrt(2 * math.sqrt(N_samples)))
print(f"The uncertainty of the standard deviation for our radiation data is {rad_uncert}. Our measured and expected variance are in statistical agreement. We did not see more variation in the data than you would expect from simple counting statistics.")
```

```python
plt.plot(time_data, temp_data)
plt.xlabel("Time (sec)")
plt.ylabel("Temperature Data (C)")
plt.title('Temperature vs Time')
```

```python
temp_mean = np.mean(np.array(temp_data))
temp_std = np.std(temp_data)
print(f"The mean is {temp_mean} and the std is {temp_std}")
```

### Part F

```python
# Since no location data, going to split the data into 5 chunks by time
chunk_1_rad = rad_data[0:20]
chunk_2_rad = rad_data[20:40]
chunk_3_rad = rad_data[40:60]
chunk_4_rad = rad_data[60:80]
chunk_5_rad = rad_data[80:99]

chunk_1_mean = np.mean(chunk_1_rad)
chunk_2_mean = np.mean(chunk_2_rad)
chunk_3_mean = np.mean(chunk_3_rad)
chunk_4_mean = np.mean(chunk_4_rad)
chunk_5_mean = np.mean(chunk_5_rad)

print(f'The mean counts from 1-19 seconds is {chunk_1_mean}. Mean counts from 20-39 seconds is {chunk_2_mean}. Mean counts from 40-59 seconds is {chunk_3_mean}. Mean counts from 60-79 seconds is {chunk_4_mean}. Mean counts from 80-99 seconds is {chunk_5_mean}. Please note for the final chunk there are only 18 data points--the total data points is 98.')
```

```python
#uncertainty on the mean = std/sqrt(N)
chunk_1_std = np.std(chunk_1_rad)
chunk_2_std = np.std(chunk_2_rad)
chunk_3_std = np.std(chunk_3_rad)
chunk_4_std = np.std(chunk_4_rad)
chunk_5_std = np.std(chunk_5_rad)

mean_unc_1 = chunk_1_std / math.sqrt(len(chunk_1_rad))
mean_unc_2 = chunk_2_std / math.sqrt(len(chunk_2_rad))
mean_unc_3 = chunk_3_std / math.sqrt(len(chunk_3_rad))
mean_unc_4 = chunk_4_std / math.sqrt(len(chunk_4_rad))
mean_unc_5 = chunk_5_std / math.sqrt(len(chunk_5_rad))

print(mean_unc_1)
print(mean_unc_2)
print(mean_unc_3)
print(mean_unc_4)
print(mean_unc_5)
```

Our first 2 "chunk" statisitcs are comparable. The next two are as well. These two paris are not comparable to each other, however. Based on the time, this second pair were measurements taken more towards the intersection of Le Roy & Hearst and then the Public Policy school. The final uncertainty on the mean measurement is the most different because it only has 18 data points as opposed to the 20 that all the other chunks had. The noticed differences aren't significant to dose levels because the difference in uncertainty on the 2 pairs that do compare is about 0.5. 

```python
rel_unc_1_per = (mean_unc_1/ chunk_1_mean) * 100
rel_unc_2_per = (mean_unc_2/ chunk_2_mean) * 100
rel_unc_3_per = (mean_unc_3/ chunk_3_mean) * 100
rel_unc_4_per = (mean_unc_4/ chunk_4_mean) * 100
rel_unc_5_per = (mean_unc_5/ chunk_5_mean) * 100
print(f'The relative uncertainty for chunk 1 is {rel_unc_1_per:.3}%.')
print(f'The relative uncertainty for chunk 2 is {rel_unc_2_per:.3}%.')
print(f'The relative uncertainty for chunk 3 is {rel_unc_3_per:.3}%.')
print(f'The relative uncertainty for chunk 4 is {rel_unc_4_per:.3}%.')
print(f'The relative uncertainty for chunk 5 is {rel_unc_5_per:.3}%.')
```

Because the relative uncertainty for each chunck is below 5%, we can conclude that the data we took was accurate.
