```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Temperature vs. humidity for outside
outside_data = pd.read_csv("Combo_Data_Outside.csv")
x0 = outside_data['Humidity']
y0 = outside_data['Temp']
```

```python
from importlib import reload
plt=reload(plt)
```

```python
plt.plot(x0, y0)
plt.xlabel('Outside Temperature')
plt.ylabel('Outside Humidity')
plt.title('Outside Temp vs. Humidity')
plt.show()
```

```python
inside_data = pd.read_csv("Combo_Data_Inside.csv")
x1 = inside_data["Humidity"]
y1 = inside_data['Temp']
plt.plot(x1, y1)
plt.xlabel('Inside Temperature')
plt.ylabel('Inside Humidity')
plt.title('Inside Temp vs. Humidity')
plt.show()
```

The hidden variable when plotting these is time because as stated the data is connected through the time measurement. For both inside and outside data, we can notice that as temperature increases, humidity increases too. This shows a positive correlation. The curve for inside shows a linear, positive line. For outside, it varies at the beginning and a slight curve is due to us walking around to get outside from inside. 

```python
x2 = inside_data["Pressure"]
y2 = inside_data['Temp']
plt.plot(x2, y2)
plt.xlabel('Inside Pressure')
plt.ylabel('Inside Temperature')
plt.title('Inside Temp vs. Pressure')
plt.show()
```

```python
x3 = inside_data["Gas"]
y3 = inside_data['Temp']
plt.plot(x3, y3)
plt.xlabel('Inside Gas')
plt.ylabel('Inside Temperature')
plt.title('Inside Temp vs. Gas')
plt.show()
```

```python
x4 = inside_data["Altitude"]
y4 = inside_data['Temp']
plt.plot(x4, y4)
plt.xlabel('Inside Altitude')
plt.ylabel('Inside Temperature')
plt.title('Inside Temp vs. Altitude')
plt.show()
```

For inside temp vs. pressure, we noticed a negative linear correlation. For inside temp vs.gas, it is not exactly linear despite having straight lines, and it is not similar to the temp vs. humidity. Inside temp vs. altitude is very similar to temp vs. humidity in that they are both positively, linearly correlated. 

```python
x5 = inside_data["PM2.5"]
y5 = inside_data['Temp']
plt.plot(y5, x5)
plt.xlabel('Inside PM 2.5')
plt.ylabel('Inside Temperature')
plt.title('Inside Temp vs. PM 2.5')
plt.show()
```

There is no correlation between temperature and PM 2.5 in this graph. We can ensure time was the same despite this non-correlation because we had both sensors on the same timescale.
