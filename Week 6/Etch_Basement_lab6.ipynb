```python
std_mean_pm1_in = np.std(x0)/ np.sqrt(300)
print(std_mean_pm1_in)

std_mean_pm1_out = np.std(y0)/ np.sqrt(300)
print("The mean uncertainty is",std_mean_pm1_out)

diff0 = np.abs(std_mean_pm1_in - std_mean_pm1_out)
#print(diff0)

ans0 = np.abs(diff0/(np.std(x0)/np.sqrt(300)))
print("The distibutions differ by", ans0, "sigmas")
```
