```python
import matplotlib.pyplot as plt

# Define building coordinates (example data)
buildings = {
    "Doe Library": (2, 4),
    "Sather Gate": (3, 2),
    "Campanile": (4, 5),
    "Evans Hall": (5, 3),
    "Memorial Stadium": (6, 1)
}

# Extract coordinates for plotting
x_coords = [coord[0] for coord in buildings.values()]
y_coords = [coord[1] for coord in buildings.values()]
labels = list(buildings.keys())

# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(x_coords, y_coords, marker='o', color='blue', label='Buildings')

# Draw lines connecting the buildings in order
plt.plot(x_coords, y_coords, color='gray', linestyle='--', linewidth=1, label='Paths')

# Add labels to the points
for i, label in enumerate(labels):
    plt.annotate(label, (x_coords[i], y_coords[i]), textcoords="offset points", xytext=(6,6), ha='left')

# Set plot title and labels
plt.title("UC Berkeley Campus Map")
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define building coordinates (example data)
buildings = {
    "Doe Library": (2, 4),
    "Sather Gate": (3, 2),
    "Campanile": (4, 5),
    "Evans Hall": (5, 3),
    "Memorial Stadium": (6, 1)
}

# Extract coordinates for plotting
x_coords = [coord[0] for coord in buildings.values()]
y_coords = [coord[1] for coord in buildings.values()]
labels = list(buildings.keys())

# Create a mesh grid for the contour background
x = np.linspace(0, 8, 100)
y = np.linspace(0, 6, 100)
X, Y = np.meshgrid(x, y)

# Simulate elevation (e.g., a mix of hills and valleys)
Z = np.sin(X) * np.cos(Y) * 10 + X + Y  # You can tweak this function

# Create the plot
plt.figure(figsize=(10, 8))

# Add contour lines
contour = plt.contourf(X, Y, Z, cmap='terrain', alpha=0.6)
plt.colorbar(contour, label='Elevation')

# Plot buildings
plt.scatter(x_coords, y_coords, marker='o', color='blue', edgecolor='black', s=100, label='Buildings')

# Draw connecting paths (optional)
plt.plot(x_coords, y_coords, color='gray', linestyle='--', linewidth=1, label='Paths')

# Add labels to the points
for i, label in enumerate(labels):
    plt.annotate(label, (x_coords[i], y_coords[i]), textcoords="offset points", xytext=(6,6), ha='left')

# Final plot touches
plt.title("UC Berkeley Campus - Topographic Map")
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```
