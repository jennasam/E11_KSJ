# Created a map that plots all of the radiation data!

```python
# average radiation at each location

hearst_rad = np.mean(hearst_rad_data)
MLK_rad = np.mean(MLK_rad_data)
north_gate_rad = np.mean(north_gate_rad_data)
binge_rad = np.mean(binge_rad_data)
doe_rad = np.mean(doe_rad_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
import matplotlib.image as mpimg

# Define building locations and radiation values
buildings_rad = {
    "Binge Coffee House": ((.5, 0.5), binge_rad),
    "Hearst Mining Hall": ((6, 8), hearst_rad),
    "MLK": ((4.45, 1.75), MLK_rad),
    "Doe": ((4.75, 4.3), doe_rad),
    "North Gate": ((4.4, 8.3), north_gate_rad)
}

# Extract coordinates and values
coords = np.array([loc for loc, _ in buildings_rad.values()])
radiation_values = np.array([val for _, val in buildings_rad.values()])
labels = list(buildings_rad.keys())

# Create interpolation grid
x = np.linspace(0, 9 , 200)
y = np.linspace(0, 10, 200)
X, Y = np.meshgrid(x, y)

# Step 5: Interpolate using 'cubic' method
Z = griddata(coords, radiation_values, (X, Y), method='cubic')

# Step 6: Plot
plt.figure(figsize=(8.5, 5))
ax = plt.gca()

#adding map image
map_img = mpimg.imread('berk_map.jpg')
plt.imshow(map_img, extent=[0, 10, 0, 10], aspect='auto')

# Contour heatmap of interpolated radiation data
contour = plt.contourf(X, Y, Z, cmap='YlGnBu', levels=100, alpha = 0.5)
plt.colorbar(contour, label='Radiation (average counts)')

# Plot building locations
x_coords = coords[:, 0]
y_coords = coords[:, 1]
plt.scatter(x_coords, y_coords, color='white', edgecolor='black', s=100, label='Sensor Locations', alpha = 0.7)

# Label each point with building name and count
for i, label in enumerate(labels):
    count = radiation_values[i]
    plt.annotate(f"{label}\n{count:.1f} counts", (x_coords[i], y_coords[i]),
                 textcoords="offset points", xytext=(6, 6), ha='left', fontsize=9)


# Final plot touches
plt.title("UC Berkeley Radiation Map (Average Counts at Each Location)")
plt.xlabel("X Coordinates")
plt.ylabel("Y Coordinates")
plt.grid(True, linestyle=':')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assume you already have these data arrays:
# hearst_rad_data, MLK_rad_data, north_gate_rad_data, binge_rad_data, doe_rad_data

# Step 1: Compute average radiation at each location
hearst = np.mean(hearst_rad_data)
MLK = np.mean(MLK_rad_data)
north_gate = np.mean(north_gate_rad_data)
binge = np.mean(binge_rad_data)
doe = np.mean(doe_rad_data)

# Step 2: Define building coordinates and radiation values
buildings = {
    "Binge Coffee House": ((2, 0.5), binge),
    "Hearst Mining Hall": ((7, 3.5), hearst),
    "MLK": ((5, 0.5), MLK),
    "Doe": ((5, 2.5), doe),
    "North Gate": ((5, 5), north_gate)
}

# Step 3: Extract coordinates and radiation values
coords = np.array([loc for loc, _ in buildings.values()])
radiation_values = np.array([val for _, val in buildings.values()])
labels = list(buildings.keys())

# Normalize radiation values for color mapping
norm = plt.Normalize(vmin=min(radiation_values), vmax=max(radiation_values))
cmap = plt.cm.plasma

# Step 4: Plotting circles with a strong ombre effect
plt.figure(figsize=(13, 9))

# Scatter plot of large circles with ombre effect
for i, (label, ((x, y), rad)) in enumerate(buildings.items()):
    size = rad * 2000  # Increase circle size for better overlap
    sc = plt.scatter(x, y,
                     s=size,  # Large circles for blending
                     color=cmap(norm(rad)),  # Color based on radiation data
                     edgecolor='black',
                     alpha=0.3,  # Lower alpha to allow blending
                     label=label)
    # Add label with the value
    plt.text(x + 0.2, y + 0.1, f"{label}\n{rad:.1f} counts", fontsize=9)

# Add a colorbar
plt.colorbar(sc, label='Radiation (average counts)')

# Final touches
plt.title("UC Berkeley Radiation Map (Ombre Effect with Blended Circles)")
plt.xlabel("X Coordinate (arbitrary campus grid)")
plt.ylabel("Y Coordinate")
plt.grid(True, linestyle=':')
plt.xlim(0, 10)
plt.ylim(0, 6)
plt.tight_layout()
plt.show()
```
