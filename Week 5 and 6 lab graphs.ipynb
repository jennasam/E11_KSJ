```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
inside_data = pd.read_csv("Combo_Data_Inside.csv")
outside_data = pd.read_csv("Combo_Data_Outside.csv")
x0 = inside_data['PM1.0']
x1 = inside_data['PM2.5']
x2 = inside_data['PM10']
x3 = inside_data['Temp']
x4 = inside_data['Gas']
x5 = inside_data['Humidity']
x6 = inside_data['Pressure']
x7 = inside_data['Altitude']

y0 = outside_data['PM1.0']
y1 = outside_data['PM2.5']
y2 = outside_data['PM10']
y3 = outside_data['Temp']
y4 = outside_data['Gas']
y5 = outside_data['Humidity']
y6 = outside_data['Pressure']
y7 = outside_data['Altitude']


plt.plot(x0, label = "time vs PM1.0")
plt.ylabel('time')
plt.xlabel('PM 1.0')
plt.legend()
plt.show()


plt.plot(x1, label = "time vs PM2.5")
plt.ylabel('time')
plt.xlabel('PM 2.5')
plt.show()

plt.plot(x2, label = "PM10")
plt.ylabel('time')
plt.xlabel('PM 10')
plt.show()

plt.plot(x3, label = "temperature")
plt.ylabel('time')
plt.xlabel('temperature')
plt.show()

plt.plot(x4, label = "Gas")
plt.ylabel('time')
plt.xlabel('Gas')
plt.show()

plt.plot(x5, label = "humidity")
plt.ylabel('time')
plt.xlabel('humidity')
plt.show()

plt.plot(x6, label = "Pressure")
plt.ylabel('time')
plt.xlabel('pressure')
plt.show()

plt.plot(x7, label = "altitude")
plt.ylabel('time')
plt.xlabel('altitude')
plt.show()
```

```python
plt.hist(x0, label = "inside PM1.0" )
plt.hist(y0, label = "outside PM1.0")
plt.ylabel('time')
plt.xlabel('PM 1.0')
plt.legend()
plt.show()


plt.hist(x1, label = "inside PM2.5")
plt.hist(y1, label = "outside PM2.5")
plt.ylabel('time')
plt.xlabel('PM 2.5')
plt.legend()
plt.show()


plt.hist(x2, label = "inside PM10")
plt.hist(y2, label = "outside PM10")
plt.ylabel('time')
plt.xlabel('PM 10')
plt.legend()
plt.show()


plt.hist(x3, label = "inside temp")
plt.hist(y3, label = "outside temp")
plt.ylabel('time')
plt.xlabel('temperature')
plt.legend()
plt.show()


plt.hist(x4, label = "inside Gas")
plt.hist(y4, label = "outside Gas")
plt.ylabel('time')
plt.xlabel('Gas')
plt.legend()
plt.show()


plt.hist(x5, label = "inside humidity")
plt.hist(y5, label = "outside humidity")
plt.ylabel('time')
plt.xlabel('humidity')
plt.legend()
plt.show()


plt.hist(x6, label = "inside pressure")
plt.hist(y6, label = "outside pressure")
plt.ylabel('time')
plt.xlabel('pressure')
plt.legend()
plt.show()


plt.hist(x6, label = "inside altitude")
plt.hist(y7, label = "outside altitude")
plt.plot(x7, label = "altitude")
plt.ylabel('time')
plt.xlabel('altitude')
plt.legend()
plt.show()

```

```python
std_mean_pm1_in = np.std(x0)/ np.sqrt(300)
print(std_mean_pm1_in)

std_mean_pm1_out = np.std(y0)/ np.sqrt(300)
print("The mean uncertainty is",std_mean_pm1_out)

diff0 = np.abs(std_mean_pm1_in - std_mean_pm1_out)
#print(diff0)

ans0 = np.abs(diff0/(np.std(x0)/np.sqrt(300)))
print("The distibutions differ by", ans0, "sigmas")
```
